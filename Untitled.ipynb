```python
target_excel = 'Summary_6174489_Glaucoma_drugs_2019.xlsx'
ref_excel = 'Verification 100 cases HKWC 2019 10 3 2022 ver 2.xlsx'

import pandas as pd

target_df = pd.read_excel(target_excel).values
ref_df = pd.read_excel(ref_excel).values
```

```python
print(target_df[:,0])
print(ref_df[:,0])

new_list = []
for x in target_df[:,0]:
    if x not in list(ref_df[:,0]):
        new_list.append(x)

print(len(target_df[:,0]))
print(len(new_list))
```

```python
import random
random_200 = random.sample(new_list, 200)

df = pd.DataFrame(random_200, columns=['Reference ID'])
df.to_excel('200_random_HKWC_2019_unduplicated.xlsx')
```
