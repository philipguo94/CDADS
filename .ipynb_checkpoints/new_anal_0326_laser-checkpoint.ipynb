```python
# total costs of all operations
OPERATION_PRICES = {'CYCLOPHOTOCOAGULATION': 31150,
                                    'TRABECULECTOM AB EXTERNO': 65600,
                                    'POSTOP REVIS SCL FISTUL':20000,
                                    'FACILIT INTRAOC CIRC NEC':20000,
                                    'GONIOSYNECHIAE LYSIS': 43700,
                                    'TRABECULOTOMY AB EXTERNO':46850}


cost_operation_year = {}
x_axis = []
for year in range(2009,2020,1):
    x_axis.append(year)
    cost = 0
    df_each_year = operations_results[pd.to_datetime(operations_results['Procedure Date (rank 1) (yyyy-mm-dd HH:MM)']).dt.year == year]
    for drug in OPERATION_PRICES.keys():
        cost += OPERATION_PRICES[drug] * len(df_each_year['Procedure Description (rank 1)'][df_each_year['Procedure Description (rank 1)']==drug])
    cost_operation_year[year] = cost
print(cost_operation_year)


fig = plt.figure(figsize=(12,10))
x_axis = cost_operation_year.keys()
no_by_operation = cost_operation_year.values()
# creating the bar plot
plt.bar(x_axis, no_by_operation, color ='maroon', width = 0.4)
 
for a,b in zip(x_axis, no_by_operation): 
    plt.text(a-0.4, b+0.3, str(b))

plt.xticks(rotation=90)
plt.xlabel("Years")
plt.ylabel("HKD")
plt.title("Total Costs of used drugs by year (all clusters)", fontsize=16)
plt.show()

```
