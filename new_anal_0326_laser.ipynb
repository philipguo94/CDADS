```python
import numpy as np

def read_operation_excel(filepath):
    total_df = []
    output_df = []
    df = pd.read_excel(filepath)
    df = df['Most Recent Procedures Performed on Attendance 1 - Procedure Description']
    return df.dropna()


def operation_filter(operations_results):
    OPERATIONS = ['IRIDOPLASTY NEC',
                                    'IRIDECTOMY NEC',
                                    'Laser Irido/coreoplasty',
                                    'Iridotomy by laser',
                                    'IRIDOTOMY W TRANSFIXION',
                                    'GONIOPUNCTURE',
                                     'CYCLOPHOTOCOAGULATION',
                                      'Trabeculoplasty by laser',
                                    'IRIDOTOMY NEC'
                 ]
    operations_results = operations_results[operations_results.isin(OPERATIONS)]
    return operations_results
```

 # Question : types and no. of surgeries by year

```python
import os

import pandas as pd

CLUSTER_NAMES = ['NTWC', 'NTEC', 'KEC', 'KWC', 'KCC', 'HKEC', 'HKWC']


root_dir = 'CDARS all filtered files/'


operation_year_dict = {}
for year in range(2009, 2020):
    operations_df = []
    
    for filename in os.listdir(root_dir):
        print(filename)
        if 'laser' in filename and str(year) in filename:
            try:
                df = read_operation_excel(root_dir + filename)
                operations_df.append(df)
            except:
                pass
    operations_results = pd.concat(operations_df)

    operation_year_dict[year] = operation_filter(operations_results)
    
```
