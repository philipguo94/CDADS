```python
import numpy as np

def read_operation_excel(filepath):
    total_df = []
    output_df = []
    df = pd.read_excel(filepath)
    total_df = [df.iloc[:, [0, 8, 13]], df.iloc[:, [0, 9, 14]],df.iloc[:, [0, 10, 15]],df.iloc[:, [0,11, 16]],df.iloc[:, [0, 12, 17]]]
    for data in total_df:
        for x in data.values:
            if not pd.isnull(x[1]) and not pd.isnull(x[2]):
                output_df.append(x)
                print(output_df[-1])
    output_df = pd.DataFrame(output_df, columns=['Reference Key', 'Procedure Description (rank 1)', 'Procedure Date (rank 1) (yyyy-mm-dd HH:MM)'])
    return output_df


def operation_filter(operations_results):
    OPERATIONS = ['CYCLOPHOTOCOAGULATION',
                                    'TRABECULECTOM AB EXTERNO',
                                    'POSTOP REVIS SCL FISTUL',
                                    'FACILIT INTRAOC CIRC NEC',
                                    'GONIOSYNECHIAE LYSIS',
                                    'TRABECULOTOMY AB EXTERNO',
                                     'TRABECULOTOMY',
                                     'GONIOTOMY']
    operations_results = operations_results[operations_results['Procedure Description (rank 1)'].isin(OPERATIONS)]
    return operations_results
```
